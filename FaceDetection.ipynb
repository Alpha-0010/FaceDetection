## Face Detection

```python
# We will build the face detection algorithm using the Sliding Window technique. Move the window throughout the 
# image and whenever a face is detected using template matching mark the face. 
# So basically for the template matching we don't need to train a model instead we can use harrCascade which
# is in-built in openCv. The creators of openCv have used 100-1000s data to train a ML model and the output of 
# that model is made into a template and they have stored this template into harrcascade.xml file.
```

```python
import cv2
import matplotlib.pyplot as plt
```

```python
img=plt.imread("./group-photo.jpg")
```

```python
plt.imshow(img)
```

```python
# Load the frontal cascade classifier
detector=cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')
```

```python
allFaces=detector.detectMultiScale(img,1.3,5)
```

```python
# The detector detected six faces and each face has four properties (x,y,w,h):-----
# x,y--> Start coordinate of the face.
# w,h--> width and height of the face.

allFaces.shape
```

```python
allFaces
```

```python
x,y,w,h = allFaces[0]
```

```python
# This is for marking one face.
img=cv2.rectangle(img,(x,y),(x+w,y+h),(0,0,0),2)
```

```python
plt.imshow(img)
```

```python
# Mark all the faces....
for face in allFaces:
    x,y,w,h = face
    img=cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
```

```python
plt.imshow(img)
```

### Detect faces in real time

```python
%matplotlib
```

```python
cam=cv2.VideoCapture(0)
```

```python
while True:
    
    # cam.read() returns two values one is the Frame and the other is ret which will be True or False depending
    # upon if the webcam is able to capture image..
    # Capture Frame-By-Frame...
    ret, frame=cam.read()
    
    if ret==False:
        continue
    
    
    allFaces=detector.detectMultiScale(frame)
    
    for face in allFaces:
        x,y,w,h = face
        frame=cv2.rectangle(frame,(x,y),(x+w,y+h),(255,0,0),2)
    
    # Display the resulting Frame..
    cv2.imshow('Face Detection',frame)
    
    # To stop the webcam if 'q' key is pressed..
    keyPressed=cv2.waitKey(1) & 0xFF
    if keyPressed== ord('q'):
        break
    

# When everything is done release the capture and destroyAllWindows..
cam.release()
cv2.destroyAllWindows()
```
